```python
import pandas as pd
```

```python
df = pd.read_csv("extracted_data.csv")
print(df.iloc[5793])
```

```python
df.rename(columns={"Date": "Date_Time"}, inplace=True)
df["Date"] = df['Date_Time'].str.split().str[:4].str.join(" ")
display(df.head())
```

```python
perfect_pattern = r'^[A-Za-z]{3},\s\d{1,2}\s[A-Za-z]{3}\s\d{4}$'
exceptions_mask = ~df['Date'].str.match(perfect_pattern, na=False)

print(f"Found {exceptions_mask.sum()} exceptions. Here they are:")
display(df[exceptions_mask][['Date_Time', 'Date']])
```
